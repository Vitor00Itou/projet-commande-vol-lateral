```python
from plane_dynamics import simulate_avion
from constants import g, Vie, gammae, psie, phie

import matplotlib.pyplot as plt
import numpy as np
```

```python
response = simulate_avion(0, 400)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle(f'Simulação com Vie={Vie:.1f} m/s e Phie={np.degrees(phie):.1f} deg')

# Velocidade
axs[0, 0].plot(response['time'], response['v'], 'tab:blue')
axs[0, 0].set_title('Velocidade (V) [m/s]')
axs[0, 0].grid(True)

# Gamma
axs[0, 1].plot(response['time'], np.degrees(response['gamma']), 'tab:orange')
axs[0, 1].set_title('Gamma (Trajetória) [deg]')
axs[0, 1].grid(True)

# Psi (Heading) - Deve mudar se Phie != 0
axs[1, 0].plot(response['time'], np.degrees(response['phi']), 'tab:green')
axs[1, 0].set_title('Psi (Proa) [deg]')
axs[1, 0].grid(True)

# Phi (Bank)
axs[1, 1].plot(response['time'], np.degrees(response['phi']), 'tab:red')
axs[1, 1].set_title('Phi (Bank) [deg]')
axs[1, 1].grid(True)
```
